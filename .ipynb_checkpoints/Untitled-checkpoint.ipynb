```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
from scipy import stats, interpolate
names = ["YY","MM","DD","hh","mm","WDIR","WSPD","GST","WVHT","DPD","APD","MWD","PRES","ATMP","WTMP","DEWP","VIS","TIDE"]
#pd.concat
df = pd.read_csv("2010.txt",delimiter ='\s+',skiprows= 2,names=names)
file_name=np.arange(2009,2023)
#Loop for importing txt files
for i in file_name:
    file = str(i) +".txt"
    df_file = pd.read_csv(file,delimiter = '\s+' , skiprows = 2,names=names)
    df = pd.concat([df,df_file])
#changing dates into datetime
df['dates'] = pd.to_datetime(df['YY']*100+df['MM'], format = '%Y%m')
#grouping colums
grouped_year_index =df.groupby(['dates'])
grouped_wind_speed_data=grouped_year_index['WSPD'].mean()
grouped_wind_pressure_data = grouped_year_index['PRES'].mean()
#slicing values
x = grouped_wind_speed_data.index
y = grouped_wind_speed_data.values
z = grouped_wind_pressure_data.values
display(df)
print(grouped_wind_speed_data)
#ploting press vs temp 
fig, axs = plt.subplots(figsize=(10,6))
axs.scatter(x,y,marker = '.',label='Wind Speed')
#setting x labels
axs.set_xlabel('Year')
axs.set_ylabel('Wind Speed(m/s)')
#setting title
axs.set_title('Annual mean wind speed')
#setting legend
fig.legend(loc='upper right')
axs.grid()
ax2 = axs.twinx()
ax2.plot(x,z,marker ='o', color ='green',label='pressure')
ax2.set_ylabel('pressure')
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
from scipy import stats, interpolate
import matplotlib.patches as mpatches
names = ["YY","MM","DD","hh","mm","WDIR","WSPD","GST","WVHT","DPD","APD","MWD","PRES","ATMP","WTMP","DEWP","VIS","TIDE"]
#pd.concat
df = pd.read_csv("2010.txt",delimiter ='\s+',skiprows= 2,names=names)
#loop for importing data 
file_name=np.arange(2009,2023)
for i in file_name:
    file = str(i) +".txt"
    df_file = pd.read_csv(file,delimiter = '\s+' , skiprows = 2,names=names)
    df = pd.concat([df,df_file])
#slicing colimns
summer=df.loc[(df['MM']>5) & (df['MM']<8),['YY','MM','hh','DD','WSPD']]
summer_grouped = summer.groupby(['YY','MM','DD'] ,as_index = False)['WSPD'].mean()
summer_grouped['datess'] = pd.to_datetime(summer_grouped['YY']*10000+summer_grouped['MM']*100+summer_grouped['DD'], format = '%Y%m%d')
display(summer_grouped)
#slicing specific year values
_2021_=summer_grouped.loc[(summer_grouped['YY']==2021),['WSPD']].values
x_2021=summer_grouped.loc[(summer_grouped['YY']==2021),['datess']].values
#calculating mean wind speed value for 2021
_2021_mean = _2021_.mean()
_2022_=summer_grouped.loc[(summer_grouped['YY']==2022),['WSPD']].values
x_2022=summer_grouped.loc[(summer_grouped['YY']==2022),['datess']].values
#calculating mean win speed value for 2022
_2022_mean = _2022_.mean()
#plotting multi columns plot 
fig,(ax3,ax4) = plt.subplots(ncols=2,figsize=(15,6))
ax3.plot(x_2021,_2021_,marker='o',color='red')
for label in ax3.get_xticklabels(which='major'):
    label.set(rotation=30, horizontalalignment='right')
ax3.set_title("2021 windspeed measurement daily")
ax3.set_xlabel('Dates')
ax3.set_ylabel('Wind Speed(Knots)')
ax3.grid()
ax3.axhline(y = _2021_mean, color = 'green', linestyle = '-') 
ax3.set_ylim([2,7])
ax4.plot(x_2022,_2022_,marker='o',color='blue')
#Loop for rotating date time dates. Found it at https://stackoverflow.com/questions/11264521/date-ticks-and-rotation
for label in ax4.get_xticklabels(which='major'):
    label.set(rotation=30, horizontalalignment='right')
#setting title
ax4.set_title("2022 windspeed measurement daily")
#setting labels
ax4.set_xlabel('Dates')
ax4.set_ylabel('Wind Speed(Knots)')
#setting grid
ax4.grid()
#setting mean value
ax4.axhline(y = _2022_mean, color = 'green', linestyle = '-') 
#setting y limit for all 6 y axis to match
ax4.set_ylim([2,7])
fig.legend(['2021','Mean wind Speed','2022'])
```

```python
_2015_=summer_grouped.loc[(summer_grouped['YY']==2015),['WSPD']].values
x_2015=summer_grouped.loc[(summer_grouped['YY']==2015),['datess']].values
#slicing 2015 mean value
_2015_meanspeed = _2015_.mean()
_2016_=summer_grouped.loc[(summer_grouped['YY']==2016),['WSPD']].values
x_2016=summer_grouped.loc[(summer_grouped['YY']==2016),['datess']].values
#slicing 2016 mean value
_2016_meanspeed = _2016_.mean()
#plotting figure
fig,(ax3,ax4) = plt.subplots(ncols=2,figsize=(15,6))
ax3.plot(x_2015,_2015_,marker='o',color='red')
#rotating x axis datetime 
for label in ax3.get_xticklabels(which='major'):
    label.set(rotation=30, horizontalalignment='right')
ax3.set_title("2015 windspeed measurement daily")
ax3.set_xlabel('Dates')
ax3.set_ylabel('Wind Speed(Knots)')
ax3.grid()
ax3.axhline(y = _2015_meanspeed, color = 'green', linestyle = '-') 
ax3.set_ylim([2,7])
ax4.plot(x_2016,_2016_,marker='o',color='blue')
#roating x axis datetime
for label in ax4.get_xticklabels(which='major'):
    label.set(rotation=30, horizontalalignment='right')
ax4.set_title("2016 windspeed measurement daily")
ax4.set_xlabel('Dates')
ax4.set_ylabel('Wind Speed(Knots)')
ax4.grid()
ax4.axhline(y = _2016_meanspeed, color = 'green', linestyle = '-')
ax4.set_ylim([2,7])
fig.legend(['2015','Mean wind Speed','2016'])
```
